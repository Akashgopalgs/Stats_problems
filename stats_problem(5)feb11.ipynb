# Cheking Skewness of data from different datasets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
valentines = pd.read_csv(r"C:\Users\akash\Documents\dataset\amazon_2024_valentines_best_sellers.csv")
valentines
```

```python
sns.distplot(valentines['price/value'],bins=30)
```

```python
print(f"mean is {valentines['price/value'].mean()}")
print(f"median is {valentines['price/value'].median()}")
from scipy import stats as st
st.mode(valentines['reviewsCount'])
```

* Amazon's valentines sales price is in Positive/Right skew
* The mean of the product price is frater than 
* Positive skew suggests that there are relatively few high-priced items pulling the mean up.
* This distribution implies that there are outliers with high prices, which affect the mean more than the median.
* Understanding this skew can help in pricing strategies and identifying influential outliers in the data

```python
wc=pd.read_csv(r"C:\Users\akash\Documents\dataset\twitchdata-update.csv")
wc
```

```python
print(f"mean   is {wc['Watch time(Minutes)'].mean()}")
print(f"median is {wc['Watch time(Minutes)'].median()}")
```

```python
sns.distplot(wc['Watch time(Minutes)'],bins=30)
```

* The mean watch time is greater than the median, indicating positive skewness.
* Positive skew suggests that there are relatively few instances of very long watch times pulling the mean up.
* This distribution implies that there are some viewers who watch for significantly longer durations, influencing the mean more than the median.
* Understanding this skew can help in content strategy, targeting viewers who engage for longer durations, and possibly detecting outliers in viewer behavior.
