```python
import pandas as pd
import numpy as np
from numpy.random import normal
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Height of Adult Humans

```python
mu = 175
sigma = 5
n = 200
```

```python
sample = normal(mu,sigma,n)
```

```python
sns.distplot(sample)
plt.title('Normal Distribution of Adult Male Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()
```

```python
first_std_upper = sample.mean() + sample.std()
first_std_upper
```

```python
first_std_lower = sample.mean() - sample.std()
first_std_lower
```

```python
second_std_upper = sample.mean() + 2* (sample.std())
second_std_upper
```

```python
second_std = sample.mean() - 2* (sample.std())
second_std
```

* First upper bounded standared deviation of adult humen's height is 179.2045936576502
* First lower bounded standared deviation of adult humen's height is 169.6729166905876
    - 60 % of datas are in First standared deviation
* Second upper bounded standared deviation of adult humen's height is 183.9704321411815
* Second lower bounded standared deviation of adult humen's height is 164.90707820705632
    - 97 % of datas are in second standared deviation

#### Errors in Measurement:
    In fields like science and engineering, measurement errors often follow a normal distribution. For example, if you repeatedly measure the length of an object with slight variations each time.

```python
mu = 10.5
sigma = 0.6
n = 25
```

```python
err = normal(mu,sigma,n)
```

```python
sns.distplot(err)
plt.title('Normal Distribution of Errors in Measurement')
plt.xlabel('Error')
plt.ylabel('Probability Density')
plt.show()
```

```python
first_std_upper = err.mean() + err.std()
first_std_upper
```

```python
first_std_lower = err.mean() - err.std()
first_std_lower
```

```python
second_std_upper = err.mean() + 2* (err.std())
second_std_upper
```

```python
second_std_lower = err.mean() - 2* (err.std())
second_std_lower
```
